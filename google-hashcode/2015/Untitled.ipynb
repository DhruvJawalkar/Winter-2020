```python
class Server():
    def __init__(self, server_id, slots, capacity):
        self.id = server_id
        self.slots = slots
        self.cap = capacity
        self.pool = -1
    
    def get_id(self):
        return self.id
    
    def get_slots_required(self):
        return self.slots
    
    def get_capacity(self):
        return self.capacity
    
    def get_pool_assigned_to(self):
        return self.pool
```
